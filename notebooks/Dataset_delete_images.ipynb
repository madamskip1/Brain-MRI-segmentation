```python
import json
import random
from PIL import Image
```

```python
images_to_delete_file = open("../project/data_modules/components/images_to_delete.json")
images_to_delete_json = json.load(images_to_delete_file)
```

```python
images_to_delete_counter = 0
for i in images_to_delete_json:
    images_to_delete_counter += len(i["files_index"])
    
display(f"Usunięto {images_to_delete_counter} zdjęć")
```

```python
display_images = 10
dirs_count = len(images_to_delete_json)
i = 0

while i < display_images:
    dir = images_to_delete_json[random.randint(0, dirs_count - 1)]
    files_count = len(dir["files_index"])
    if files_count > 0:
        file_index = dir["files_index"][random.randint(0, files_count - 1)]
        display(dir["dir"] + "_" + str(file_index))
        image = Image.open(f"../dataset/kaggle_3m/{dir['dir']}/{dir['dir']}_{file_index}.tif")
        display(image)
        channels = image.split()
        display(channels[0])
        display(channels[1])
        display(channels[2])
        display("__________________________________________")
        i += 1
```
